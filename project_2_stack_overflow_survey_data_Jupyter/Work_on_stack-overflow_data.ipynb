```python
df["LanguageHaveWorkedWith"]
```

```python
df["LanguageHaveWorkedWith"]=df["LanguageHaveWorkedWith"].fillna("N/A")
```

```python
l3=df["LanguageHaveWorkedWith"].to_list()
l3
```

```python
l4=[]
for items in l3:
    if ";" in items:
        x=items.split(";")
        l4.extend(x)
    else:
        l4.append(items)
```

```python
l4
```

```python
language=pd.Series(l4)
language.value_counts()
```

```python
df5=pd.DataFrame(language.value_counts())
df5.reset_index(inplace=True)
df5.rename(columns={"index":"language",0:"No of people uses"},inplace=True)
df5.head()
```

# 10.Make a df containing the languages people uses and their popularity

```python
df5["popularity in percentage"]=(df5["No of people uses"]/df5["No of people uses"].sum())*100
df5
```

```python
# df5.to_csv("language popularity.csv")
```

# 11. Make a bar graph of 10 most popular languges

```python
plt.ylabel("languages")
plt.xlabel("popularity(%)")
plt.title("Languages vs Popularity(%) Graph")
plt.barh(df5["language"].iloc[0:10],df5["popularity in percentage"].iloc[0:10])
plt.tight_layout()
plt.legend()
# plt.savefig("Languages vs Popularity graph.jpg")
plt.show()
```
