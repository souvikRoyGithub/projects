# 2.Make a column(description) joining two columns(gender,age) with a keyword "and"

```python
df["Description"]=df["Age"]+" and "+df["Gender"]
```

# 3.Drop columns(age,gender)

```python
df.drop(columns=["Age","Gender"],inplace=True)
df.columns
```

```python
df.loc[:,"Description"].fillna("na and na",inplace=True)
```

```python
l=(df.loc[:,"Description"]).to_list()
l
```

```python
l1=[]
l2=[]
for item in l:
    x=item.split(" and ")
    l1.append(x[0])
    l2.append(x[1])
```

```python
l1
```

```python
l2
```

```python
df.head()
```

# 4.Spliting column(description) as columns(gender,age)

```python
df["Age"]=pd.Series(l1)
df["Gender"]=pd.Series(l2)
df.columns
```
