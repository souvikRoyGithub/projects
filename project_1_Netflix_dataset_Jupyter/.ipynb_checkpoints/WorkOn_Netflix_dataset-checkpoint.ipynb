```python
l
```

```python
l1=[]
for i in l:
    i=i.replace('[','')
    i=i.replace(']','')
    i=i.replace('\'','')
    i=i.replace(' ','')
    l1.append(i)
print(l1)
```

```python
l2=[]
for i in l1:
    if "," in i:
        x=i.split(",")
        l2.extend(x)
    else:
        l2.append(i)
l2
        
```

```python
l2.count("crime")
```

```python
edited_genres=pd.Series(l2)
edited_genres
```
