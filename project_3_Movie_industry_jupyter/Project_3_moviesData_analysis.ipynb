```python
# looking at the percentage of missing data in each column
```

```python
for col in df.columns:
    print(f"nan value percentage of column '{col}' is: {np.mean(df[col].isnull())}")
```

```python
# droping the rows in which having the null values of budget and gross.
```

```python
df=df.dropna(axis=0,how='any',subset=["budget","gross"])
```

```python
# changing the dataTypes of columns:budget and gross
```

```python
df["budget"]=df["budget"].astype(int)
df["gross"]=df["gross"].astype(int)
df.dtypes
```

```python
df
```

```python
# 'released' year and 'year' should be the same year although its not the case, have to correct the 'year' column
```

```python
df.loc[:,["released","year"]]
```

```python
df.drop(columns="year",inplace=True)
```

```python
df.columns
```

```python
temp_list=[]
for i in df["released"]:
    index=int(i.rfind(',')+1)
    i=i[index:index+5]
    temp_list.append(i)
df.insert(4,"year",temp_list,True)
```
